## Ex 01

$$\max_{W_2 \in [0,W_1]}u(W_1-W_2)$$

If the individual's utility function is an increasing funtion with consumption, the optimal choice for the individual is to eat all the cake in the first time period, i.e. $W_2=0$.

## Ex 02

The maximum problem in period 2:
$$\max_{W_3 \in [0,W_2]} u(W_2-W_3)$$
The maximum problem in period 1:
$$\max_{W_2 \in [0,W_1]} u(W_1-W_2)+\beta \max_{W_3 \in [0,W_2]} u(W_2-W_3)$$

## Ex 03

In period 3:
$$\max_{W_4\in[0,W_3]}u(W_3-W_4)$$
In period 2:
$$\max_{W_3\in[0,W_2]}u(W_2-W_3)+\beta\max_{W_4\in[0,W_3]}u(W_3-W_4)$$
In period 1:
$$\max_{W_2\in[0,W_1]}u(W_1-W_2)+\beta\max_{W_3\in[0,W_2]}u(W_2-W_3)+\beta^2\max_{W_4\in[0,W_3]}u(W_3-W_4)$$

To optimize the utility in period 3, the individual needs to eat all the remaing cake ($W_3$) in period3, so $W_4=0$.
Then apply $W_4=0$ to the maximum problem in period 2. From the first order condition, we can get that $W_3=\frac{9}{19}W_2$.
Then apply this equation to the first order condition in period 1. We finally get that $W_2=\frac{171}{272}$, $W_3=\frac{81}{272}$, $W_4=0$.

The evolving process is shown below.

## Ex 04

In period T-1:
$$\max_{W_T}u(W_{T-1}-W_T)+\beta u(W_T)$$

The first order condition is:
$$-u^\prime(W_{T-1}-W_T)+\beta u^\prime(W_T)=0$$
where
$$W_T=\psi_{T-1}(W_{T-1}).$$

Thus, the function $V_{T-1}$ becomes
$$V_{T-1}(W_{T-1})=u(W_{T-1}-\psi_{T-1}(W_{T-1}))+\beta u(\psi_{T-1}(W_{T-1})).$$

## Ex 05

When $u(c)=\ln(c)$, in period T, it needs to satisfies:
$$W_{T+1}=\psi_T(W_T))=0$$
$$V_T(W_T)=u(W_T)=\ln(W_T).$$

And the first order condition in period T-1 becomes:
$$\frac{-1}{W_{T-1}-W_T}+\beta\frac{1}{W_T}=0.$$

Thus,
$$W_T=\psi_{T-1}(W_{T-1})=\frac{\beta}{1+\beta}W_{T-1}$$
$$V_{T-1}(W_{T-1})=\ln(W_{T-1}-\frac{\beta}{1+\beta}W_{T-1})+\beta\ln(\frac{\beta}{1+\beta}W_{T-1}).$$

As a result,
$$V_T(\bar{W})=\ln(\bar{W})$$
$$V_{T-1}(\bar{W})=\ln(\frac{1}{1+\beta}\bar{W})+\beta\ln(\frac{\beta}{1+\beta}\bar{W})$$
$$\psi_T(\bar{W}))=0$$
$$\psi_{T-1}(\bar{W}))=\frac{\beta}{1+\beta}\bar{W}.$$
So $V_T(\bar{W})\neq V_{T-1}(\bar{W})\quad$ and $\quad\psi_T(\bar{W}))\neq \psi_{T-1}(\bar{W}))$.

## Ex 06

The Bellman equation at time T-2 is:
$$V_{T-2}(W_{T-2})=\max_{W_{T-1}}\ln(W_{T-2}-W_{T-1})+\beta V_T(W_{T-1}).$$

From Exercise 05, we know that $V_{T-1}(W_{T-1})=\ln(\frac{1}{1+\beta}W_{T-1})+\beta\ln(\frac{\beta}{1+\beta}W_{T-1}).\quad$Applying it to the first order condition of the function above, we can get:
$$\frac{-1}{W_{T-2}-W_{T-1}}+\frac{\beta}{W_{T-1}}+\frac{\beta^2}{W_{T-1}}=0.$$
$$\Rightarrow W_{T-1}=\psi_{T-2}(W_{T-2})=\frac{\beta^2+\beta}{\beta^2+\beta+1}W_{T-2}.$$

Thus,
$$V_{T-2}(W_{T-2})=\ln(\frac{W_{T-2}}{\beta^2+\beta+1})+\beta\ln(\frac{\beta W_{T-2}}{\beta^2+\beta+1})+\beta^2\ln(\frac{\beta^2W_{T-2}}{\beta^2+\beta+1}).$$

## Ex 07

Guess that 
$$V_{T-s}(W_{T-s})=\ln(\frac{W_{T-s}}{\beta^s+\beta^{s-1}+\dots+1})+\dots+\beta^s\ln(\frac{\beta^sW_{T-s}}{\beta^s+\beta^{s-1}+\dots+1})$$

It is easy to see that when s=1 or s=2 equation above is satisfied. Assume that it hodes when s=k. Then for s=k+1:
$$V_{T-k-1}(W_{T-k-1})=\max_{W_{T-k}}\ln(W_{T-k-1}-W_{T-k})+\beta V_{T-k}(W_{T-k}).$$
First order condition is:
$$\frac{1}{W_{T-k-1}-W_{T-k}}=(\beta+\dots+\beta^{k+1})W_{T-k}$$
$$\Rightarrow W_{T-k}=\psi_{T-k-1}(W_{T-k-1})=\frac{\beta+\dots+\beta^{k+1}}{\beta+\dots+\beta^{k+1}+1}W_{T-k-1}$$

Then we can get that
$$\begin{split} V_{T-k-1}(W_{T-k-1})&=\ln(\frac{W_{T-k-1}}{\beta+\dots+\beta^{k+1}+1})+\beta V_{T-k}(W_{T-k})\\
&=\ln(\frac{W_{T-k-1}}{\beta+\dots+\beta^{k+1}+1})+\dots+\beta^{k+1}\ln(\frac{\beta^{k+1}W_{T-k-1}}{\beta+\dots+\beta^{k+1}+1}) \end{split}$$
The induction is completed; thus, we conclude that 
$$V_{T-s}(W_{T-s})=\ln(\frac{W_{T-s}}{\beta^s+\beta^{s-1}+\dots+1})+\dots+\beta^s\ln(\frac{\beta^sW_{T-s}}{\beta^s+\beta^{s-1}+\dots+1})$$
$$\psi_{T-s}(W_{T-s})=\frac{\beta^s+\beta^{s-1}+\dots+\beta}{\beta^s+\beta^{s-1}+\dots+\beta+1}W_{T-s}$$

When s tends to infinity,since $0<\beta<1$, we can get that
$$\lim_{s \to \infty}\psi_{T-s}(W_{T-s})=\beta W_{T-s}$$
$$\lim_{s \to \infty} V_{T-s}(W_{T-s})=\frac{\ln((1-\beta)W_{T-s})}{1-\beta}+\frac{\beta}{(1-\beta)^2}\ln(\beta).$$

## Ex 10

```python
def u(c):
    return np.log(c)
```

```python
Wprime_vec = W_vec.reshape(1, N)
W_mat = np.tile(W_vec, (1, N))
Wprime_mat = np.tile(Wprime_vec, (N, 1))
C_mat = W_mat - Wprime_mat
c_pos = C_mat<=0
C_mat[c_pos] = 1e-10
util = u(C_mat)
```

```python
beta = 0.9
Vprime = np.zeros_like(util)
T_tml = (u(C_mat) + beta * Vprime).max(axis=1).reshape(N, 1)
Wprime_idx = np.argmax(u(C_mat) + beta * Vprime, axis=1)
```

```python
sd = 0.5
mean = 4*0.5
M = 7
epsilon_vec = np.linspace(mean-3*sd, mean+3*sd, M)
```

```python
from scipy.stats import norm
def gamma(x):
    return norm(loc=mean, scale=sd).pdf(x)
```

```python
gamma_vec = gamma(epsilon_vec)
plt.plot(epsilon_vec, gamma_vec)
```

## Ex 17

```python
# use the w_vec, wprime_vec, C_mat in previous problems
V_T_plus1 = np.zeros((N, M))
U_mat = u(C_mat)
cube = np.array([U_mat*e for e in epsilon_vec])
ev = V_T_plus1 @ gamma_vec # N*1
ev_mat = np.tile(ev.reshape(1, N), (N, 1)) 
ev_mat[c_pos] = -9e+4
ev_cube = np.array([ev_mat for e in epsilon_vec])
V_new = np.zeros((N, M))
Wprime = np.zeros((N, M))
V_T = cube + beta * ev_cube
for i in range(N):
    array = V_T[:, i, :]
    idx = np.argmax(array, axis=1)
    Wprime[i] = W_vec.reshape(1,N)[0][idx]
    V_new[i] = array.max(axis=1)
```
